```python
# Series Operations:- 
import pandas as pd

x = [0, 1, 2, 3, 4, 5, 7]
y = [6, 7, 8, 9, 5]

s5 = pd.Series(x)
s6 = pd.Series(y)
s5
s6
```

```python
s5.add(s6)
```

```python
s5.sub(s6)
```

```python
s7 = s5.mul(s6)
```

```python
s5.div(s6)
```

```python
print('median', s7.median())
print('max', s7.max())
print('min', s7.min())
```

```python
# Visualization in Pandas
import numpy as np
%matplotlib inline

ts  = pd.Series(np.random.randn(50), index = pd.date_range('today', periods = 50))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(50, 4), index = ts.index, columns = ['A', 'B', 'X', 'Y'])
df = df.cumsum()
df.plot()
```
